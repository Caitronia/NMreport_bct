Curve Fitting: y=ax+b 

```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])

n = len(x)

a = (n*np.sum(x*y)-np.sum(x)*np.sum(y))/(n*np.sum(x*x)-np.sum(x)**2)
b = (np.sum(y)-a*np.sum(x))/n

print("Equation: y =", round(a,4), "x +", round(b,4))
```

Curve Fitting: y=a+bx

```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])

n = len(x)

b = (n*np.sum(x*y)-np.sum(x)*np.sum(y))/(n*np.sum(x*x)-np.sum(x)**2)
a = (np.sum(y)-b*np.sum(x))/n

print("Equation: y =", round(a,4), "+", round(b,4), "x")
```

Curve Fitting: y=ab^x

```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([2,4,8,16,32])

Y = np.log(y)
n = len(x)

B = (n*np.sum(x*Y)-np.sum(x)*np.sum(Y))/(n*np.sum(x*x)-np.sum(x)**2)
A = (np.sum(Y)-B*np.sum(x))/n

a = np.exp(A)
b = np.exp(B)

print("Equation: y =", round(a,4), "*", round(b,4), "^x")
```

Lagrange Interpolation

```python
def lagrange(x, y, xp):
    n = len(x)
    yp = 0

    for i in range(n):
        p = 1
        for j in range(n):
            if i != j:
                p *= (xp - x[j]) / (x[i] - x[j])
        yp += p * y[i]

    return yp

x = [1,2,3,4]
y = [1,4,8,16]
xp = 2.5

print("Interpolated value =", lagrange(x,y,xp))
```

Simpson's 1/3 Rule

```python
def f(x):
    return x*x

a = 0
b = 4
n = 4

h = (b-a)/n

s = f(a)+f(b)

for i in range(1,n):
    if i%2==0:
        s += 2*f(a+i*h)
    else:
        s += 4*f(a+i*h)

ans = h*s/3

print("Integral =", ans)
```

Weddle's Rule

```python
def f(x):
    return x*x

a = 0
b = 6

h = (b-a)/6

x = [a+i*h for i in range(7)]

ans = (3*h/10)*(f(x[0])+5*f(x[1])+f(x[2])+6*f(x[3])+f(x[4])+5*f(x[5])+f(x[6]))

print("Integral =", ans)
```

Composite Trapezoidal Rule

```python
def f(x):
    return x*2

a = 0
b = 4
n = 4

h = (b-a)/n

s = f(a)+f(b)

for i in range(1,n):
    s += 2*f(a+i*h)

ans = h*s/2

print("Integral =", ans)
```

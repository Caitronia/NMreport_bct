## 1. Lagrange Interpolation

```
def lagrange_interpolation(x, y, x_val):
    n = len(x)
    result = 0
    for i in range(n):
        term = y[i]
        for j in range(n):
            if j != i:
                term *= (x_val - x[j]) / (x[i] - x[j])
        result += term
    return result

x = [5, 7, 11, 13]
y = [150, 392, 1452, 2366]
x_val = 9
interp = lagrange_interpolation(x, y, x_val)
print(f"Interpolated value at x = {x_val} is {interp:.4f}")
```

## 2. Lagrange Interpolation for y = ab^x

```
import math

def lagrange_exponential_interpolation(x, y, x_val):
    n = len(x)
    Y = [math.log(val) for val in y]
    result_Y = 0
    for i in range(n):
        term = Y[i]
        for j in range(n):
            if j != i:
                term *= (x_val - x[j]) / (x[i] - x[j])
        result_Y += term
    return math.exp(result_Y)

x = [1, 2, 3]
y = [2.0, 8.0, 32.0]
x_val = 1.5
interp = lagrange_exponential_interpolation(x, y, x_val)
print(f"Interpolated value at x = {x_val} is {interp:.4f}")
```

## 3. Lagrange Interpolation for y = a + bx

```
def lagrange_linear_interpolation(x, y, x_val):
    n = len(x)
    result = 0
    for i in range(n):
        term = y[i]
        for j in range(n):
            if j != i:
                term *= (x_val - x[j]) / (x[i] - x[j])
        result += term
    return result

x = [1, 2, 3]
y = [3.0, 5.0, 7.0]
x_val = 1.5
interp = lagrange_linear_interpolation(x, y, x_val)
print(f"Interpolated value at x = {x_val} is {interp:.4f}")
```

## 4. Simpson's 1/3 and 3/8 Rules

```
def simpsons_13(f, a, b, n):
    if n % 2 != 0:
        raise ValueError("For Simpson's 1/3 Rule, 'n' must be an even integer.")
    h = (b - a) / n
    integration_sum = f(a) + f(b)
    for i in range(1, n):
        x = a + i * h
        if i % 2 == 0:
            integration_sum += 2 * f(x)
        else:
            integration_sum += 4 * f(x)
    return (h / 3) * integration_sum

def simpsons_38(f, a, b, n):
    if n % 3 != 0:
        raise ValueError("For Simpson's 3/8 Rule, 'n' must be a multiple of 3.")
    h = (b - a) / n
    integration_sum = f(a) + f(b)
    for i in range(1, n):
        x = a + i * h
        if i % 3 == 0:
            integration_sum += 2 * f(x)
        else:
            integration_sum += 3 * f(x)
    return (3 * h / 8) * integration_sum

if __name__ == "__main__":
    import math
    func = math.sin
    lower_limit = 0
    upper_limit = math.pi
    n_13 = 100
    n_38 = 99
    result_13 = simpsons_13(func, lower_limit, upper_limit, n_13)
    result_38 = simpsons_38(func, lower_limit, upper_limit, n_38)
    print(f"Exact value: 2.0")
    print(f"Simpson's 1/3 Result (n={n_13}): {result_13:.10f}")
    print(f"Simpson's 3/8 Result (n={n_38}): {result_38:.10f}")
```

## 5. Boole's and Weddle's Rules

```
def booles_rule(f, a, b, n):
    if n % 4 != 0:
        raise ValueError("For Boole's Rule, the number of intervals 'n' must be a multiple of 4.")
    h = (b - a) / n
    integration_sum = 0
    for i in range(0, n, 4):
        x0 = a + i * h
        x1 = a + (i + 1) * h
        x2 = a + (i + 2) * h
        x3 = a + (i + 3) * h
        x4 = a + (i + 4) * h
        integration_sum += (7 * f(x0) + 32 * f(x1) + 12 * f(x2) + 32 * f(x3) + 7 * f(x4))
    return (2 * h / 45) * integration_sum

def weddles_rule(f, a, b, n):
    if n % 6 != 0:
        raise ValueError("For Weddle's Rule, the number of intervals 'n' must be a multiple of 6.")
    h = (b - a) / n
    integration_sum = 0
    for i in range(0, n, 6):
        x0 = a + i * h
        x1 = a + (i + 1) * h
        x2 = a + (i + 2) * h
        x3 = a + (i + 3) * h
        x4 = a + (i + 4) * h
        x5 = a + (i + 5) * h
        x6 = a + (i + 6) * h
        integration_sum += (f(x0) + 5 * f(x1) + f(x2) + 6 * f(x3) + f(x4) + 5 * f(x5) + f(x6))
    return (3 * h / 10) * integration_sum

if __name__ == "__main__":
    import math

    def my_func(x):
        return 1 / (1 + x**2)

    lower_limit = 0
    upper_limit = 6
    n_boole = 12
    n_weddle = 12
    result_boole = booles_rule(my_func, lower_limit, upper_limit, n_boole)
    result_weddle = weddles_rule(my_func, lower_limit, upper_limit, n_weddle)
    exact_value = math.atan(upper_limit) - math.atan(lower_limit)
    print(f"Exact Value: {exact_value:.12f}")
    print(f"Boole's Rule Result  (n={n_boole}): {result_boole:.12f}")
    print(f"Weddle's Rule Result (n={n_weddle}): {result_weddle:.12f}")
```

## 6. Composite Trapezoidal Rule

```
def composite_trapezoidal(f, a, b, n):
    if n < 1:
        raise ValueError("The number of intervals 'n' must be at least 1.")
    h = (b - a) / n
    integration_sum = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        x = a + i * h
        integration_sum += f(x)
    return h * integration_sum

if __name__ == "__main__":
    import math

    def my_func(x):
        return x**2

    lower_limit = 0
    upper_limit = 3
    subintervals = 100
    result = composite_trapezoidal(my_func, lower_limit, upper_limit, subintervals)
    exact_value = 9.0
    print(f"Exact Value: {exact_value:.6f}")
    print(f"Trapezoidal Result (n={subintervals}): {result:.6f}")
    print(f"Absolute Error: {abs(exact_value - result):.6e}")
```

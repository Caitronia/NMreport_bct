# Exponential Curve Fitting

### y = ab^x

```python
import math

n = int(input("Enter number of data points: "))

sum_x = 0
sum_Y = 0
sum_x2 = 0
sum_xY = 0

for i in range(n):
    x = float(input(f"x[{i+1}] = "))
    y = float(input(f"y[{i+1}] = "))

    Y = math.log10(y)

    sum_x += x
    sum_Y += Y
    sum_x2 += x * x
    sum_xY += x * Y

B = (n * sum_xY - sum_x * sum_Y) / (n * sum_x2 - sum_x ** 2)
A = (sum_Y - B * sum_x) / n

a = 10 ** A
b = 10 ** B

print("\nBest Fit Exponential Curve:")
print(f"y = {a:.4f} * ({b:.4f})^x")
```

# Linear Curve Fitting

### y = a + bx

```python
n = int(input("Enter number of data points: "))

sum_x = 0
sum_y = 0
sum_xy = 0
sum_x2 = 0

for i in range(n):
    x = float(input(f"x[{i+1}] = "))
    y = float(input(f"y[{i+1}] = "))

    sum_x += x
    sum_y += y
    sum_xy += x * y
    sum_x2 += x ** 2

b = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
a = (sum_y - b * sum_x) / n

print("\nBest Fit Straight Line:")
print(f"y = {a:.4f} + {b:.4f}x")
```

# Lagrange Interpolation

```python
def lagrange_interpolation(x, y, x_value):
    n = len(x)
    result = 0

    for i in range(n):
        term = y[i]

        for j in range(n):
            if i != j:
                term *= (x_value - x[j]) / (x[i] - x[j])

        result += term

    return result

n = int(input("Enter number of data points: "))

x = []
y = []

for i in range(n):
    x.append(float(input(f"x[{i+1}] = ")))
    y.append(float(input(f"y[{i+1}] = ")))

x_value = float(input("Enter interpolation point: "))

answer = lagrange_interpolation(x, y, x_value)

print(f"\nInterpolated value at x = {x_value} is {answer:.4f}")
```

# Simpson's 1/3 Rule

```python
def f(x):
    return x ** 2

a = float(input("Enter lower limit: "))
b = float(input("Enter upper limit: "))
n = int(input("Enter even number of intervals: "))

if n % 2 != 0:
    print("Number of intervals must be even.")
else:
    h = (b - a) / n

    total = f(a) + f(b)

    for i in range(1, n):
        x = a + i * h

        if i % 2 == 0:
            total += 2 * f(x)
        else:
            total += 4 * f(x)

    integral = (h / 3) * total

    print("Integral =", integral)
```

# Boole's Rule

```python
def f(x):
    return x ** 2

a = float(input("Enter lower limit: "))
b = float(input("Enter upper limit: "))

h = (b - a) / 4

integral = (2 * h / 45) * (
    7 * f(a)
    + 32 * f(a + h)
    + 12 * f(a + 2 * h)
    + 32 * f(a + 3 * h)
    + 7 * f(b)
)

print("Integral =", integral)
```

# Weddle's Rule

```python
def f(x):
    return x ** 2

a = float(input("Enter lower limit: "))
b = float(input("Enter upper limit: "))

h = (b - a) / 6

integral = (3 * h / 10) * (
    f(a)
    + 5 * f(a + h)
    + f(a + 2 * h)
    + 6 * f(a + 3 * h)
    + f(a + 4 * h)
    + 5 * f(a + 5 * h)
    + f(b)
)

print("Integral =", integral)
```

# Composite Trapezoidal Rule

```python
def f(x):
    return x ** 2

a = float(input("Enter lower limit: "))
b = float(input("Enter upper limit: "))
n = int(input("Enter number of intervals: "))

h = (b - a) / n

total = f(a) + f(b)

for i in range(1, n):
    total += 2 * f(a + i * h)

integral = (h / 2) * total

print("Integral =", integral)
```

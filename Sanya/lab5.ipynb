**1. y=ab^x**

```python
import numpy as np

x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2.5, 4.2, 7.1, 12.0, 20.3], dtype=float)

Y = np.log(y)

n = len(x)
sum_x = np.sum(x)
sum_Y = np.sum(Y)
sum_xY = np.sum(x * Y)
sum_x2 = np.sum(x ** 2)

# Least squares formula
B = (n * sum_xY - sum_x * sum_Y) / (n * sum_x2 - sum_x ** 2)
A = (sum_Y - B * sum_x) / n

a = np.exp(A)
b = np.exp(B)

print("a =", a)
print("b =", b)

print(f"\nFitted equation: y = {a:.4f} * ({b:.4f})^x")
```

**2. y=ax+b**

```python
import numpy as np

x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 5, 4, 5], dtype=float)

n = len(x)

sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x ** 2)

a = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
b = (sum_y - a * sum_x) / n

print("a =", a)
print("b =", b)

print(f"\nBest fit equation: y = {a:.4f}x + {b:.4f}")
```

**3. Lagrange interpolation**

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

xp = float(input("Enter the value of x: "))

n = len(x)
yp = 0

for i in range(n):
    p = 1
    for j in range(n):
        if i != j:
            p = p * (xp - x[j]) / (x[i] - x[j])
    yp = yp + p * y[i]

print("Interpolated value at x =", xp, "is", yp)
```

**4. Simpson's 3/8 th Rule**

```python
def f(x):
   return x**3 + 2*x + 1  

a = 0
b = 3
n = int(input("Enter number of subintervals (multiple of 3): "))

if n % 3 != 0:
    print("Error: Number of subintervals must be a multiple of 3.")
else:
    h = (b - a) / n
    s = f(a) + f(b)

    for i in range(1, n):
        x = a + i * h
        if i % 3 == 0:
            s += 2 * f(x)
        else:
            s += 3 * f(x)

    result = (3 * h / 8) * s

    print("Approximate value of the integral =", result)
```

**5. Weddle's Rule**

```python
def f(x):
    return x**3 + 2*x + 1   

a = 0
b = 6
n = int(input("Enter number of subintervals (multiple of 6): "))

if n % 6 != 0:
    print("Error: Number of subintervals must be a multiple of 6.")
else:
    h = (b - a) / n
    result = 0

    for i in range(0, n, 6):
        x0 = a + i * h
        x1 = x0 + h
        x2 = x0 + 2 * h
        x3 = x0 + 3 * h
        x4 = x0 + 4 * h
        x5 = x0 + 5 * h
        x6 = x0 + 6 * h

        result += (3 * h / 10) * (
            f(x0) + 5*f(x1) + f(x2) + 6*f(x3)
            + f(x4) + 5*f(x5) + f(x6)
        )

    print("Approximate value of the integral =", result)
```

**6. Trapezoidal Rule**

```python
def f(x):
    return x**3 + 2*x + 1 
a = 0
b = 5
n = int(input("Enter number of subintervals: "))

h = (b - a) / n

s = f(a) + f(b)

for i in range(1, n):
    x = a + i * h
    s += 2 * f(x)

result = (h / 2) * s

print("Approximate value of the integral =", result)
```

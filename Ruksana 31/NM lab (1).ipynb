## Lagrange interpolation 


### 1. y=ab^x

```python
def lagrange_interpolation(x, y, x_val):
    n = len(x)
    result = 0

    for i in range(n):
        term = y[i]

        for j in range(n):
            if j != i:
                term *= (x_val - x[j]) / (x[i] - x[j])

        result += term

    return result


x = [5, 7, 11, 13]
y = [150, 392, 1452, 2366]
x_val = 9

interp = lagrange_interpolation(x, y, x_val)

print(f"Interpolated value at x = {x_val} is {interp:.4f}")
```

### i) simpson 1/3 rules

```python
def f(x):
    return x**2     


def simpson_one_third(a, b, n):
    h = (b - a) / n
    s = f(a) + f(b)

    for i in range(1, n):
        if i % 2 == 0:
            s += 2 * f(a + i * h)
        else:
            s += 4 * f(a + i * h)

    return (h / 3) * s


a = 0
b = 4
n = 4     

result = simpson_one_third(a, b, n)
print("Integral =", result)
```

### ii) Composite of simpson 1/3 

```python
def f(x):
    return x**2     

def composite_simpson_one_third(a, b, n):
    h = (b - a) / n
    s = f(a) + f(b)

    for i in range(1, n):
        x = a + i * h
        if i % 2 == 0:
            s += 2 * f(x)
        else:
            s += 4 * f(x)

    return (h / 3) * s


a = 0
b = 4
n = 4    

result = composite_simpson_one_third(a, b, n)
print("Integral =", result)
```

### iii) simpson 3/8 rules

```python
def f(x):
    return x**2      


def simpson_three_eighth(a, b, n):
    h = (b - a) / n
    s = f(a) + f(b)

    for i in range(1, n):
        if i % 3 == 0:
            s += 2 * f(a + i * h)
        else:
            s += 3 * f(a + i * h)

    return (3 * h / 8) * s


a = 0
b = 3
n = 3     

result = simpson_three_eighth(a, b, n)
print("Integral =", result)
```

### iv)  Composite simpson 3/8 rules

```python
def f(x):
    return x**2     

def composite_simpson_three_eighth(a, b, n):
    h = (b - a) / n
    s = f(a) + f(b)

    for i in range(1, n):
        x = a + i * h
        if i % 3 == 0:
            s += 2 * f(x)
        else:
            s += 3 * f(x)

    return (3 * h / 8) * s


a = 0
b = 6
n = 6     

result = composite_simpson_three_eighth(a, b, n)
print("Integral =", result)
```

### Weddle's rule 

```python
def f(x):
    return x**2     

def weddle_rule(a, b):
    h = (b - a) / 6

    result = (3 * h / 10) * (
        f(a)
        + 5 * f(a + h)
        + f(a + 2 * h)
        + 6 * f(a + 3 * h)
        + f(a + 4 * h)
        + 5 * f(a + 5 * h)
        + f(b)
    )

    return result


a = 0
b = 6

result = weddle_rule(a, b)
print("Integral =", result)
```

### Boolee's rule 

```python
def f(x):
    return x**2      
def boole_rule(a, b):
    h = (b - a) / 4

    result = (2 * h / 45) * (
        7 * f(a)
        + 32 * f(a + h)
        + 12 * f(a + 2 * h)
        + 32 * f(a + 3 * h)
        + 7 * f(b)
    )

    return result

a = 0
b = 4

result = boole_rule(a, b)
print("Integral =", result)
```

### Composite Trapezoidal Rule

```python
def f(x):
    return x**2      
def composite_trapezoidal(a, b, n):
    h = (b - a) / n
    s = f(a) + f(b)

    for i in range(1, n):
        x = a + i * h
        s += 2 * f(x)

    return (h / 2) * s


a = 0
b = 4
n = 4

result = composite_trapezoidal(a, b, n)
print("Integral =", result)
```

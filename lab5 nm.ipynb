1. Curve Fitting: y=ax+b (Straight Line)

```python
import numpy as np

# Input data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 5, 7, 10, 12])

# Least Squares Method
n = len(x)

a = (n*np.sum(x*y) - np.sum(x)*np.sum(y)) / (n*np.sum(x*x) - (np.sum(x))**2)
b = (np.sum(y) - a*np.sum(x))/n

print("Equation of best fit:")
print(f"y = {a:.4f}x + {b:.4f}")
```

3. Curve Fitting: y=ab^x

```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([2,4,8,16,32])

Y = np.log10(y)

n = len(x)

B = (n*np.sum(x*Y)-np.sum(x)*np.sum(Y))/(n*np.sum(x*x)-(np.sum(x))**2)
A = (np.sum(Y)-B*np.sum(x))/n

a = 10**A
b = 10**B

print(f"y = {a:.4f}({b:.4f})^x")
```

4. Lagrange Interpolation

```python
def lagrange(x, y, xp):
    n = len(x)
    yp = 0

    for i in range(n):
        p = 1

        for j in range(n):
            if i != j:
                p *= (xp - x[j])/(x[i]-x[j])

        yp += p*y[i]

    return yp


x = [1,2,3]
y = [1,4,9]

xp = 2.5

print("Interpolated value =", lagrange(x,y,xp))
```

5. Simpson's 1/3 Rule

```python
def f(x):
    return x**2

a = 0
b = 4
n = 4

h = (b-a)/n

s = f(a)+f(b)

for i in range(1,n):
    if i%2==0:
        s += 2*f(a+i*h)
    else:
        s += 4*f(a+i*h)

ans = h*s/3

print("Integral =", ans)
```

6. Weddle's Rule

```python
def f(x):
    return x**2

a = 0
b = 6

h = (b-a)/6

ans = (3*h/10)*(f(a)+5*f(a+h)+f(a+2*h)+6*f(a+3*h)+f(a+4*h)+5*f(a+5*h)+f(b))

print("Integral =", ans)
```

7. Composite Trapezoidal Rule

```python
def f(x):
    return x**2

a = 0
b = 4
n = 4

h = (b-a)/n

s = f(a)+f(b)

for i in range(1,n):
    s += 2*f(a+i*h)

ans = h*s/2

print("Integral =", ans)
```
